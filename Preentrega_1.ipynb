Se definieron las columnas del dataset, siendo cada una de ellas una característica definida. A continuación se agrega una lista de las definiciones de cada característica, de acuerdo a la [documentación del dataset.](https://www.kaggle.com/datasets/salvatorerastelli/spotify-and-youtube/data?select=Spotify_Youtube.csv)

- **Track**: nombre de la canción, como se ve en la plataforma Spotify.
- **Artist**: nombre del artista.
- **Url_spotify**: url de acceso al perfil de Spotify del artista.
- **Album**: el álbum en el que está contenida la canción en Spotify.
- **Album_type**: indica si la canción se lanza en Spotify como single o está incluida en un álbum.
- **Uri**: Un enlace de Spotify utilizado para encontrar la canción a través de la API.
- **Danceability**: describe qué tan adecuada es una pista para bailar basándose en una combinación de elementos musicales que incluyen el tempo, la estabilidad del ritmo, la fuerza del ritmo y la regularidad general. Un valor de 0.0 es el menos bailable y 1.0 es el más bailable.
- **Energy**: es una medida de 0.0 a 1.0 y representa una medida perceptiva de intensidad y actividad. Normalmente, las pistas enérgicas se sienten rápidas, ruidosas y ruidosas. Por ejemplo, el death metal tiene mucha energía, mientras que un preludio de Bach obtiene una puntuación baja en la escala. Las características de percepción que contribuyen a este atributo incluyen rango dinámico, volumen percibido, timbre, velocidad de inicio y entropía general.
- **Key**: la clave en la que se encuentra la pista. Los números enteros se asignan a los tonos utilizando la notación estándar de clase de tono. P.ej. 0 = C, 1 = C♯/D♭, 2 = D, y así sucesivamente. Si no se detectó ninguna clave, el valor es -1.
- **Loudness**: el volumen general de una pista en decibeles (dB). Los valores de sonoridad se promedian en toda la pista y son útiles para comparar el volumen relativo de las pistas. El volumen es la cualidad de un sonido que es el principal correlato psicológico de la fuerza física (amplitud). Los valores suelen oscilar entre -60 y 0 db.
- **Speechiness**: detecta la presencia de palabras habladas en una pista. Cuanto más exclusivamente hablada sea la grabación (por ejemplo, un programa de entrevistas, un audiolibro, poesía), más cercano a 1,0 será el valor del atributo. Los valores superiores a 0,66 describen pistas que probablemente estén compuestas exclusivamente de palabras habladas. Los valores entre 0.33 y 0.66 describen pistas que pueden contener música y voz, ya sea en secciones o en capas, incluidos casos como la música rap. Los valores inferiores a 0.33 probablemente representen música y otras pistas que no sean de voz.
- **Acousticness**: una medida de confianza de 0.0 a 1.0 sobre si la pista es acústica. 1.0 representa una alta confianza en que la pista es acústica.
- **Instrumentalness**: predice si una pista no contiene voces. Los sonidos "Ooh" y "aah" se tratan como instrumentales en este contexto. Las pistas de rap o de palabra hablada son claramente "vocales". Cuanto más cerca esté el valor de instrumentalidad de 1.0, mayor será la probabilidad de que la pista no contenga contenido vocal. Los valores superiores a 0,5 pretenden representar pistas instrumentales, pero la confianza es mayor a medida que el valor se acerca a 1.0.
- **Liveness**: detecta la presencia de una audiencia en la grabación. Los valores de vida más altos representan una mayor probabilidad de que la pista se haya interpretado en vivo. Un valor superior a 0.8 proporciona una gran probabilidad de que la pista esté activa.
- **Valence**: una medida de 0.0 a 1.0 que describe la positividad musical transmitida por una pista. Las pistas con valencia alta suenan más positivas (por ejemplo, felices, alegres, eufóricas), mientras que las pistas con valencia baja suenan más negativas (por ejemplo, tristes, deprimidas, enojadas).
- **Tempo**: el tempo general estimado de una pista en latidos por minuto (BPM). En terminología musical, el tempo es la velocidad o el ritmo de una pieza determinada y se deriva directamente de la duración promedio del tiempo.
- **Duration_ms**: la duración de la pista en milisegundos.
- **Stream**: cantidad de streams de la canción en Spotify.
- **Url_youtube**: URL del video vinculado a la canción en Youtube, si lo tiene.
- **Title**: título del videoclip en Youtube.
- **Channel**: nombre del canal de Youtube que publicó ese video.
- **Views**: cantidad de vistas.
- **Likes**: cantidad de "me gusta".
- **Comments**: cantidad de comentarios.
- **Description**: descripción del video en Youtube.
- **Licensed**: indica si el video representa contenido con licencia, lo que significa que el contenido se subió a un canal vinculado a un socio de contenido de YouTube y luego ese socio lo reclamó.
- **official_video**: valor booleano que indica si el video encontrado es el video oficial de la canción.

Existe una característica, **Unnamed:0**, generada automáticamente para dar a cada observación un índice.

##Data cleaning
En pasos previos se identificaron que hay valores ausentes. En esta sección se definirá la eliminación o imputación de éstos.

```python
#Resolución de valores ausentes

#Si se eliminan
print("La mayor cantidad de observaciones a perder es 576, correspodiente a datos ausentes en la característica Stream.")
total_observaciones= 20718
max_observaciones_perdidas=576
porcentaje_observaciones_a_perder = max_observaciones_perdidas/total_observaciones*100
print("Si se eliminan las filas con datos ausentes, el porcentaje de observaciones perdidas sobre el total es %" + str(porcentaje_observaciones_a_perder))

#dataset sin esos valores ausentes
dataset_limpio = dataset.dropna()
print("Al eliminar las observaciones afectadas por valores ausentes, el dataset queda con 19170 observaciones")
dataset_limpio.shape


```

```python
#Carga de bibliotecas para realizar gráficos
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Columnas a mantener
columnas =["Danceability", "Stream", "Energy", "Views", "Duration_ms"]
subset_dataset = dataset[columnas]
subset_dataset_limpio = dataset_limpio[columnas]

print("Dataset con valores ausentes")
#Histogramas de dataset con valores ausentes

num_cols = len(subset_dataset.columns)
num_rows = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 6))

for i, column in enumerate(subset_dataset.columns):
    row_index = i // num_cols
    col_index = i % num_cols
    subset_dataset[column].plot.hist(ax=axes[row_index, col_index], bins=15, alpha=0.7, color='blue', edgecolor='black')
    axes[row_index, col_index].set_title(f'Histograma de {column}')
    axes[row_index, col_index].set_xlabel(column)
    axes[row_index, col_index].set_ylabel('Frecuencia')

fig.suptitle('Histogramas de dataset', fontsize=16)


plt.tight_layout()
plt.show()

print("Estadística descriptiva")
dataset.describe()



```

```python
print("Dataset sin valores ausentes")
#Histogramas de dataset sin valores ausentes

num_cols = len(subset_dataset.columns)
num_rows = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 6))

for i, column in enumerate(subset_dataset.columns):
    row_index = i // num_cols
    col_index = i % num_cols
    subset_dataset[column].plot.hist(ax=axes[row_index, col_index], bins=15, alpha=0.7, color='purple', edgecolor='black')
    axes[row_index, col_index].set_title(f'Histograma de {column}')
    axes[row_index, col_index].set_xlabel(column)
    axes[row_index, col_index].set_ylabel('Frecuencia')

fig.suptitle('Histogramas de dataset limpio', fontsize=16)


plt.tight_layout()
plt.show()

print("Estadística descriptiva")
dataset_limpio.describe()
```

No se observa que la eliminación de las observaciones con datos ausentes afecten las medias, medianas y otros valores estadísticos así como la forma de los histogramas, por lo que en los próximos pasos se utilizará el dataset limpio sin NaN.

## Exploración de datos a través de gráficos

### Top 10 de canciones en Spotify tendrán alta bailabilidad

```python
top10_tracks_data = dataset_limpio.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)

top10_tracks_sumtotal = top10_tracks_data.sum()

colores = sns.light_palette("green", n_colors=10, reverse=True)
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x=top10_tracks_data.values, y=top10_tracks_data.index, ax=ax, palette=colores)

sns.despine(left=True)

# Adicionar a soma das visualizações dos TOP 10 ao título do eixo X
ax.set_xlabel(f'Total de reproducciones en Spotify: {top10_tracks_sumtotal} streams', fontsize=10)
plt.title('Top 10 de canciones de Spotify', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
top10_tracks_data_y = dataset_limpio.groupby('Track')['Views'].sum().sort_values(ascending=False).head(10)

top10_tracks_sumtotal_y = top10_tracks_data.sum()

colores = sns.light_palette("red", n_colors=10, reverse=True)
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x=top10_tracks_data_y.values, y=top10_tracks_data_y.index, ax=ax, palette=colores)

sns.despine(left=True)

# Adicionar a soma das visualizações dos TOP 10 ao título do eixo X
ax.set_xlabel(f'Total de reproducciones en Youtube: {top10_tracks_sumtotal_y} views', fontsize=10)
plt.title('Top 10 de canciones de Youtube', fontsize=15)

plt.tight_layout()
plt.show()
```

### Las canciones singles tienen mayor cantidad de reproducciones que el resto

```python
album_type_count = dataset_limpio['Album_type'].value_counts()
labels = album_type_count.index.tolist()
sizes = album_type_count.values.tolist()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Categoría de album')
plt.legend(labels, loc='right')

plt.show()
```

```python
df_singles = dataset_limpio[dataset_limpio['Album_type'] == 'single']

df_albumes= dataset_limpio[(dataset_limpio['Album_type'] == 'album') | (dataset_limpio['Album_type'] == 'compilation')]

top10_tracks_singles = df_singles.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)

top10_tracks_not_singles = df_albumes.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)

# Convert series to DataFrames and add a column to identify the type
top10_tracks_singles_df = top10_tracks_singles.reset_index()
top10_tracks_singles_df['Type'] = 'Single'

top10_tracks_not_singles_df = top10_tracks_not_singles.reset_index()
top10_tracks_not_singles_df['Type'] = 'Album/Compilation'

# Concatenar dataframes para poder graficar
combined_df = pd.concat([top10_tracks_singles_df, top10_tracks_not_singles_df])

# Barplot
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x='Stream', y='Track', hue='Type', data=combined_df, palette='viridis', ax=ax)

# Titulos y etiquetas
ax.set_title('Top 10 Tracks en Spotify (Singles vs. Albums/Compilations)', fontsize=16)
ax.set_xlabel('Streams', fontsize=12)
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
top10_tracks_singles = df_singles.groupby('Track')['Views'].sum().sort_values(ascending=False).head(10)

top10_tracks_not_singles = df_albumes.groupby('Track')['Views'].sum().sort_values(ascending=False).head(10)

# Convert series to DataFrames and add a column to identify the type
top10_tracks_singles_df = top10_tracks_singles.reset_index()
top10_tracks_singles_df['Type'] = 'Single'

top10_tracks_not_singles_df = top10_tracks_not_singles.reset_index()
top10_tracks_not_singles_df['Type'] = 'Album/Compilation'

# Concatenar dataframes para poder graficar
combined_df = pd.concat([top10_tracks_singles_df, top10_tracks_not_singles_df])

# Barplot
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x='Views', y='Track', hue='Type', data=combined_df, palette='viridis', ax=ax)

# Titulos y etiquetas
ax.set_title('Top 10 Tracks en Youtube (Singles vs. Albums/Compilations)', fontsize=16)
ax.set_xlabel('Views', fontsize=12)
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

Contrario a la hipótesis propuesta, las canciones parte de albumes tiene mayor cantidad de reproducciones tanto en Youtube como en Spotify.

### Las canciones más cortas tienen mayor cantidad de reproducciones


```python
top10_tracks_cortos = dataset_limpio.groupby('Track')['Duration_ms'].sum().sort_values(ascending=False).head(10)
print("Top 10 de canciones más largas")
print(top10_tracks_cortos)

top10_tracks_largos = dataset_limpio.groupby('Track')['Duration_ms'].sum().sort_values(ascending=True).head(10)
print("Top 10 de canciones más cortas")
print(top10_tracks_largos)
```

```python
tracks_largos = ['52 Non Stop Dilbar Dilbar Remix(Remix By Kedrock,Sd Style)', 'El Ultimo Adiós - Varios Artistas Version', 'High Rated Gabru 52 Non Stop Hits(Remix By Mandy Birgi,Birgi Veerz)', 'Poesia Acústica 13', 'Color Esperanza 2020', 'Ocean Waves for Sleep', 'Set dos Casados', 'Resistiré', 'Om Bhurbhuwasah', 'Poesia Acústica 10: Recomeçar']

tracks_cortos = ['Steven Universe Future (Opening Theme) / Title Card', 'Gentle Piano Melodies', 'Soothing Baby White Nosie', 'White Noise Gentle Piano', 'Peacefull Paino and White Noise', 'Bedtime White Noise and Piano Tunes', 'Relaxing Baby White Noise Piano', 'Deep Sleep White Noise For Babies', 'Magic Piano and White Noise', 'Soothing Piano and White Nosie']

dataset_tracks_cortos = dataset_limpio[dataset_limpio['Track'].isin(tracks_cortos)]

dataset_tracks_largos = dataset_limpio[dataset_limpio['Track'].isin(tracks_largos)]


top10_tracks_cortos = dataset_tracks_cortos.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)

top10_tracks_largos = dataset_tracks_largos.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)

# Convert series to DataFrames and add a column to identify the type
top10_tracks_cortos_df = top10_tracks_cortos.reset_index()
top10_tracks_cortos_df['Type'] = 'Cortos'

top10_tracks_largos_df = top10_tracks_largos.reset_index()
top10_tracks_largos_df['Type'] = 'Largos'

# Concatenar dataframes para poder graficar
combined_df_duration = pd.concat([top10_tracks_cortos_df, top10_tracks_largos_df])

# Barplot
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x='Stream', y='Track', hue='Type', data=combined_df_duration, palette='viridis', ax=ax)

# Titulos y etiquetas
ax.set_title('Top 10 Tracks en Spotify (Cortos vs. Largos)', fontsize=16)
ax.set_xlabel('Streams', fontsize=12)
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

Se realizaron los top 10 de las canciones de mayor longitud como de menor longitud tanto en Spotify como en Youtube. En ambos casos, se puede observar claramente que en ambas plataformas las canciones de mayor longitud tienen muchas más reproducciones que las canciones más cortas
